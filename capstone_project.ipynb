```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the Excel file
df = pd.read_csv('Sales Dataset.csv')

# Calculate the average purchase amount for each product category
avg_purchase_amount = df.groupby('ProductCategory')['PurchaseAmount'].mean()

# Visualize the distribution of purchase amounts across different product categories
plt.figure(figsize=(10, 6))
avg_purchase_amount.plot(kind='bar')
plt.title('Average Purchase Amount by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Average Purchase Amount')
plt.show()

# Find the product category with the highest and lowest purchase amounts
max_category = avg_purchase_amount.idxmax()
min_category = avg_purchase_amount.idxmin()
print(f'Product category with the highest purchase amount: {max_category}')
print(f'Product category with the lowest purchase amount: {min_category}')
```
